```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_reviews = pd.read_csv('googleplaystore_user_reviews.csv')
df_reviews.head()
```

```python
# Check for Nulls and Data Types

df_reviews.info()
df_reviews.isnull().sum()
```

```python
# Drop Completely Empty or Irrelevant Rows

df_reviews.dropna(how='all', inplace=True)  # Drop rows where all values are NaN
```

```python
# Drop rows with any missing values

df_reviews.dropna(inplace=True)
```

```python
# Remove Duplicates

df_reviews.drop_duplicates(inplace=True)
```

```python
# Ensure Correct Data Types

df_reviews['Sentiment_Polarity'] = pd.to_numeric(df_reviews['Sentiment_Polarity'], errors='coerce')
df_reviews['Sentiment_Subjectivity'] = pd.to_numeric(df_reviews['Sentiment_Subjectivity'], errors='coerce')
```

```python
# Normalize Text

df_reviews['Translated_Review'] = df_reviews['Translated_Review'].astype(str).str.lower()
df_reviews['Translated_Review'] = df_reviews['Translated_Review'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)
```

```python
# Export the Cleaned Dataset

df_reviews.to_csv('googleplaystore_user_reviews_cleaned.csv', index=False)
files.download('googleplaystore_user_reviews_cleaned.csv')
```
